```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv('automobile_data.csv')
```

#Engine-size

```python
plt.figure(figsize=(8,5))
```

```python
sns.histplot(data['engine-size'],bins = 20, kde=True)
plt.xlabel('Engine size')
plt.ylabel('Frequency')
plt.title('Histogram of Engine size')
plt.show()
```

#Insights 1
Distribution: The histogram reveals that the engine sizes are not uniformly distributed and are instead right-skewed.

Common Range: Most engine sizes fall between 100 and 200 units, indicating this is the typical range for cars in the dataset.

Rare Large Engines: Engine sizes greater than 300 are rare and likely belong to high-performance or specialized vehicles.

Smooth Density Curve: The KDE curve overlays the histogram and highlights the peak around 120-150 units.

```python
plt.figure(figsize=(8,5))
sns.boxplot(x=data['engine-size'])
plt.title('Box Plot of Engine Size')
plt.xlabel('Engine Size')
plt.show()
```

#Insights 2
Median Value: The central line in the box shows the median engine size, which is close to the lower end of the distribution, reinforcing the histogram insight.

Interquartile Range (IQR): The majority of engine sizes are concentrated within a narrow range (the box), indicating consistency in engine sizes for most vehicles.

Outliers: A few significant outliers (beyond the whiskers) represent vehicles with exceptionally large engine sizes.

```python
plt.figure(figsize=(12,6))
```

```python
sns.countplot(y='make', data=data, order=data['make'].value_counts().index)
plt.title('Count Plot of Car Makes')
plt.xlabel('Count')
plt.ylabel('Car make')
plt.show()
```

#Insights 3
Toyota Leads: Toyota has the highest frequency in this dataset. This suggests stronger representation in the sample, though this may not reflect overall market share.

Mid-Tier Representation: Brands like Nissan, Mazda, and Mitsubishi are also prominent, reflecting their significant market share but slightly lower representation compared to Toyota.

Rare Brands: Makes like Alfa-Romeo, Isuzu, Renault, and Mercury have minimal representation, suggesting they cater to niche markets or have lower popularity in the regions covered by the dataset.

#Line Plot: Price vs. Symboling

```python
plt.figure(figsize=(6,4))
sns.lineplot(data=data, x='symboling', y='price')
plt.title('Line plot: Price vs Symboling')
plt.xlabel("symboling")
plt.ylabel("Price")
plt.show()
```

#Insights 4

Trend: Prices generally drop for symboling levels around 1 but rise significantly for both lower (-2) and higher (3) symboling levels.

Interpretation: Symboling might correlate with risk or premium nature of vehicles, with extreme levels indicating either specialty cars or outliers.

Non-linear Relationship:

The plot shows a non-linear relationship between symboling and price. Prices decrease sharply around symboling level 1 and then increase dramatically for higher symboling levels.

#Scatter Plot: Engine Size vs. Price

```python
plt.figure(figsize=(6,3))
sns.scatterplot(data=data, x='engine-size', y='price')
plt.xlabel("Engine-size")
plt.ylabel("Price")
plt.title("Scatter-Plot: Engine Size vs price")
plt.show()
```

#Insights 5

Positive Correlation: Larger engine sizes generally correspond to higher prices, though the relationship is not perfectly linear.

Clusters: A dense cluster of affordable cars is observed for engine sizes between 100-150 units.

#Bar Plot: Average Price by Car Make

```python
plt.figure(figsize=(12, 6))
avg_price_by_make = data.groupby('make')['price'].mean().sort_values(ascending=False)
sns.barplot(x=avg_price_by_make.values, y=avg_price_by_make.index, palette='rainbow')
plt.title('Bar Plot: Average Price by Car make')
plt.xlabel('Average Price')
plt.ylabel('Car Make')
plt.show()
```

#Insight 6

**High-End Brands:**Jaguar, Mercedes-Benz, and Porsche have the highest average prices, suggesting their vehicles cater to premium and luxury markets.

Economical Brands: Chevrolet, Dodge, and Plymouth are on the lower end, likely representing budget-friendly vehicles.

#Clustered Bar Plot: Price by Fuel Type and Aspiration

```python
plt.figure(figsize=(10,6))
sns.barplot(data=data, x='fuel-type',y='price', hue='aspiration')
plt.xlabel('Fuel type')
plt.ylabel('Price')
plt.title('Clustered Bar Plot: Price by Fuel Type and Aspiration')
plt.show()
```

#Insights 7

Observation: Turbocharged diesel cars have the highest prices, reflecting their efficiency and premium features.

Gas-Powered Cars: Standard aspiration gas cars have lower prices compared to their turbocharged counterparts.

#Box Plot: Price Distribution by Body Style

```python
plt.figure(figsize=(10,6))
sns.boxplot(data=data, x='body-style', y='price')
plt.title('Box Plot: Price Distribution by Body style')
plt.xlabel('Body Style')
plt.ylabel('Price')
plt.show()
```

#Insights 8

Convertible and Hardtop Cars: Have higher median prices, indicating they might belong to premium categories.

Hatchbacks: Tend to be more affordable, with the lowest median prices and minimal variability.
